## **ğŸ“Œ 3. Funciones y RecursiÃ³n en Python**
### **Â¿QuÃ© es la recursiÃ³n?**  
La **recursiÃ³n** es una tÃ©cnica donde una funciÃ³n se llama a sÃ­ misma para resolver un problema en partes mÃ¡s pequeÃ±as.  

ğŸ“Œ **Ejemplo bÃ¡sico de recursiÃ³n en Python:**  
```python
def cuenta_regresiva(n):
    if n <= 0:
        print("Â¡Boom!")
    else:
        print(n)
        cuenta_regresiva(n - 1)

cuenta_regresiva(5)
```

ğŸ”¹ **ExplicaciÃ³n:**  
1. La funciÃ³n se llama a sÃ­ misma con `n - 1` hasta llegar a `n = 0`.  
2. En cada llamada, **se ejecuta una parte del problema**.

## **ğŸ“Œ 4. Dibujando una Estrella Fractal con RecursiÃ³n**  
ğŸ“Œ **Â¿QuÃ© hace esta funciÃ³n?**  
- **Si el tamaÃ±o de la estrella es pequeÃ±o, se detiene.**  
- **Si no, dibuja una estrella grande y luego estrellas mÃ¡s pequeÃ±as en cada punta.**  

ğŸ“Œ **CÃ³digo Explicado:**
```python
def star(tortuga, size):
    if size <= 10:
        return
    else:
        tortuga.begin_fill()
        for i in range(5):
            tortuga.forward(size)
            star(tortuga, size / 3)  # Llamado recursivo para estrellas mÃ¡s pequeÃ±as
            tortuga.left(216)
        tortuga.end_fill()
```

ğŸ”¹ **ExplicaciÃ³n:**  
1. **Caso base:** Si `size <= 10`, la funciÃ³n se detiene.  
2. **Bucle `for` de 5 lados:** Dibuja una estrella grande.  
3. **Llamado recursivo:** Llama a `star(tortuga, size / 3)` para hacer una estrella mÃ¡s pequeÃ±a en cada punta.  
4. **Ãngulo de giro `216Â°`**: Forma correcta de una estrella de 5 puntas.  

ğŸ“Œ **Llamado de la funciÃ³n:**  
```python
star(bob, 360)  # Dibuja una estrella grande
```

## **ğŸ“Œ 6. CÃ³digo Completo Paso a Paso**

```python
import turtle

# ConfiguraciÃ³n de la pantalla
bob = turtle.Turtle()
bob.getscreen().bgcolor("#994444")
bob.penup()
bob.goto((-200, 100))
bob.pendown()

# FunciÃ³n recursiva para dibujar una estrella fractal
def star(tortuga, size):
    if size <= 10:
        return
    else:
        tortuga.begin_fill()
        for i in range(5):
            tortuga.forward(size)
            star(tortuga, size / 3)
            tortuga.left(216)
        tortuga.end_fill()

# Dibujar la estrella
star(bob, 360)

# Crear otra tortuga para escribir el mensaje
mensaje = turtle.Turtle()
mensaje.color("yellow")
mensaje.penup()
mensaje.goto(0, -250)
mensaje.hideturtle()

# Escribir el mensaje con tamaÃ±o grande
mensaje.write("ğŸŒŸ Â¡BIENVENIDOS, CAMPISTAS DE PROGRAMACIÃ“N! ğŸš€", align="center", font=("Arial", 30, "bold"))

turtle.done()
```
